```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sensor_data.csv')
data['Time'] = pd.to_datetime(data['Time'])  # Convert the 'Time' column to datetime
data.head()  # Display the first few rows of the DataFrame
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Time'], data['Temperature'], label='Temperature (C)')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Time'], data['Humidity'], label='Humidity (%)', color='orange')
plt.xlabel('Time')
plt.ylabel('Humidity (%)')
plt.title('Humidity Over Time')
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature (C)', color='tab:blue')
ax1.plot(data['Time'], data['Temperature'], label='Temperature (C)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Humidity (%)', color='tab:orange')
ax2.plot(data['Time'], data['Humidity'], label='Humidity (%)', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

fig.tight_layout()  # To ensure the right y-label is not slightly clipped
plt.title('Temperature and Humidity Over Time')
plt.show()
```
